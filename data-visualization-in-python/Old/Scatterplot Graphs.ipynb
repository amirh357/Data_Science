```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.plot(x='sepal_length', y='sepal_width', kind='scatter')
```

```python
plt.figure(figsize=(12,7.5))
sns.regplot(x=iris['sepal_length'],y=iris['petal_length'], fit_reg=False);
plot1=sns.regplot(x=iris['sepal_length'],y=iris['petal_length'], fit_reg=False).get_figure()
plot1.savefig('scatter_color.png')
```

```python
plt.rcParams['figure.figsize']=(16,10)
sns.lmplot( x='sepal_length', y='petal_length', data=iris, fit_reg=False, legend=True)
plt.savefig('scatter_no_color.png')
```

```python
plt.rcParams['figure.figsize']=(16,10)
sns.lmplot( x='sepal_length', y='petal_length', data=iris, hue='species', fit_reg=False, legend=True)
plt.savefig('scatter_color.png')
```

```python
colors={'versicolor':'r','setosa':'g','virginica':'b'}
plt.scatter(x=iris.sepal_length, y=iris.petal_length, c=iris['species'].apply(lambda x: colors[x]));
```
