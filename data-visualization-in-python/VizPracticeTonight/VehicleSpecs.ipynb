# Vehicle Specifications Worksheet

```python
#Import Dependencies
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline


#Jupyter magic
```

```python
#you can import excel sheets into pandas... sometimes.
vehicles = pd.read_excel('./vehicle.xlsx')
vehicles.head()
```

```python
#What's the shape of your data
vehicles.shape
```

```python
#How many unique values do you have in the characters feature?
vehicles['characters'].unique().shape[0]
```

```python
#How many unique kart styles do you have for the characters?
vehicles.groupby('characters').describe(include = ['object']).kart
```

```python
#how many unique tires?
vehicles.groupby('characters').describe(include = ['object']).tires
```

```python
#how many unique gliders?
vehicles.groupby('characters').describe(include = ['object']).glider
```

```python
#what is the distribution of speeds - show numerically
vehicles.speed.describe()
```

```python
#what is the distribution of speeds - show visually
plt.boxplot('speed',data = vehicles);
plt.xlabel('speed')
plt.ylabel('values')
```

```python
#is there a correlation between weight and accelleration for the top
#20 combinations with the highest accelleration? Show numerically
vehicles.sort_values(by = 'acceleration',ascending = False,inplace = True)
vehicles.head(20)
```
